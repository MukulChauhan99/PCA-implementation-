```python
from keras.datasets import mnist

(X_train,Y_train),(X_test,Y_test)=mnist.load_data()

X=X_test
Y=Y_test

X=X_test.reshape(-1,28*28)
```

```python

from sklearn.preprocessing import StandardScaler
model=StandardScaler()
X=model.fit_transform(X)
X.shape
```

```python

from sklearn.decomposition import PCA

model1=PCA(n_components=3)
z_=-1*model1.fit_transform(X)

z_
```

```python

import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt


df=pd.DataFrame(z_,columns=['PCA1','PCA2','PCA3'])
df2=pd.DataFrame(Y,columns=['label'])

new=pd.concat([df,df2],axis=1)
```

```python
plt.figure(figsize=(20,10))
ax = plt.axes(projection ="3d")
fg=sns.FacetGrid(new,hue='label',height=10)
fg.map(ax.scatter3D,"PCA1","PCA2","PCA3")
fg.add_legend()
plt.show()
```
